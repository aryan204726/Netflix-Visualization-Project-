```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv("netflix_titles.csv")


```

```python
# cleaning the data

df = df.dropna(subset=["type","country","rating","duration"])

```

```python
# 
type_count = df["type"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_count.index,type_count.values,color = ["blue","orange"])
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Movies_VS_TvShows.png")
plt.show()
```

```python
ratings_count = df["rating"].value_counts()
plt.figure(figsize = (8,6))
plt.pie(ratings_count , labels = ratings_count.index, autopct = "%1.1f%%", startangle = 90)
plt.title("Percentage of Contant Rating %")
plt.tight_layout()
plt.savefig("Content_Ratings_pie.jpg")
plt.show()
```

```python
movies_df = df[df["type"] == "Movie"].copy()
movies_df["duration_int"] = movies_df["duration"].str.replace("min"," ").astype(int)
plt.figure(figsize=(8,6))
plt.hist(movies_df["duration_int"] , bins = 30 , color = "purple" , edgecolor = "black")
plt.title("Distribution of Movies duration")
plt.xlabel("Duration(minutes)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.savefig("Movie_duration_hist.png")
plt.show()
```

```python
release_counts = df["release_year"].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index , release_counts.values , color = "red")
plt.title("Realease Year VS NO. of Shows")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.tight_layout()
plt.grid(True)
plt.savefig("Release_Year_Scatter.png")
plt.show()
```

```python
country_count = df["country"].value_counts().head(10)
plt.figure(figsize= (8,6))
plt.barh(country_count.index , country_count.values , color ="teal" )
plt.title("Top 10 country by number of shows ")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top10_countries.png")
plt.show()
```

```python
content_by_year = df.groupby(["release_year" , "type"]).size().unstack().fillna(0)
fig , ax = plt.subplots(1,2 , figsize = (12,5))

ax[0].plot(content_by_year.index , content_by_year["Movie"], color = "blue")
ax[0].set_title("Movies released Per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("No. of Movies")


ax[1].bar(content_by_year.index , content_by_year["TV Show"] , color = "green")
ax[1].set_title("TV show released Per Year")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("No. of TV Shows")

fig.suptitle("Comparision of Movies and Tv shows released over year")
plt.tight_layout()
plt.savefig("Movie_TvShows_compar.png")
plt.show()
```
